```python
from nba_api.stats.static import players, teams
from nba_api.stats.endpoints import shotchartdetail
import pandas as pd
```

```python
df_teams = pd.DataFrame(teams.get_teams())

df_teams
```

```python
%%time

import time

for season in range(2007, 2023):
    print(season)
    
    l = []
    for index, (team) in enumerate(df_teams['id']):
        print(f'{index}: {team}')
        
        l.append(shotchartdetail.ShotChartDetail(
            team_id=team, 
            player_id=0, 
            season_type_all_star=['Regular Season', 'Playoffs'], 
            season_nullable=f'{season}-{str(season+1)[2:]}', 
            context_measure_simple='FGA'
        ).get_data_frames()[0])
        
        time.sleep(0.7)
    
    df = pd.concat(l, ignore_index=True)
    df.insert(0, 'season', season)
    df.to_csv(f'./data/shot_data_{season}.csv', index=False)

'Done'
```
